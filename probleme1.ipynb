```python
import json
import pandas as pd
```

```python
def json2dataframe(file_path:str):
    print("Started Reading JSON file")
    with open(file_path, "r") as read_file:
        print("Converting JSON encoded data into Python dictionary")
        data = json.load(read_file)
    return data
```

```python
data = json2dataframe(file_path='OCR/10611.json')
```

```python
data
```

```python
def get_vertices(data):   
    vertices = {}
    pages = (data.get('pageFiles'))
    if len(pages.keys()) > 1 :
        for i in len(pages.keys()):
            vertices[i] = pages.get(i).get("vertices")
    else:
        vertices = pages.get("1").get("vertices")
    
    return pd.json_normalize(vertices)
```

```python
vertices = get_vertices(data)
```

```python
vertices
```

```python
def get_lines(vertices, alpha=.5):
    lines = []
    for y_line in vertices["gravityCenter.y"].unique():
        line = ""
        for i in range(len(vertices["gravityCenter.y"])):
            y = vertices["gravityCenter.y"][i]
            if y == y_line or y == y_line+alpha or y == y_line-alpha :
                line = line +" "+ vertices["text"][i]
        lines.append(line)
    lines = pd.DataFrame(lines)
    lines.columns =['lines']
    return lines
```

```python
lines = get_lines(vertices)
```
